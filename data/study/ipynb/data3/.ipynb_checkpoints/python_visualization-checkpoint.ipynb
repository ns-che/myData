```python
import pandas as pd
import numpy as np

np.random.seed(42)

dates = pd.date_range('2024-01-01', '2024-12-31')

regions = ['서울', '부산', '대구', '광주', '인천']
products = ['A', 'B','C','D']

data = {
    'date': np.random.choice(dates, 1000),
    'region': np.random.choice(regions, 1000),
    'product': np.random.choice(products, 1000),
    'sales': np.random.randint(100, 1000, 1000),
    'profit': np.random.randint(10, 200, 1000)
}
df = pd.DataFrame(data)

df = df.sort_values('date').reset_index(drop=True)

df.to_csv('sales_data.csv', index=False, encoding='utf-8-sig')
print(df.head())
```

```python
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rc('font', family='Malgun Gothic')   # 한글 폰트 설정
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 깨짐 방지

mean_classgroup_df = df.groupby('class')['mean'].mean().reset_index()

# x축 라벨 매핑
x_labels = {1: "Class1", 2: "Class2"}

fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # subplot 2개

# 첫 번째 subplot
sns.barplot(data=mean_classgroup_df, x='class', y='mean',
            hue='class', palette='dark', alpha=0.6, ax=axes[0])
axes[0].set_xticks(range(len(mean_classgroup_df)))
axes[0].set_xticklabels([x_labels[i] for i in mean_classgroup_df['class']])
axes[0].set_xlabel("")
axes[0].set_ylabel("점수")
axes[0].legend_.remove()  # 범례 제거
axes[0].set_title("평균 점수 비교")
axes[0].set_ylim(0, 100)

for p in axes[0].patches[:len(mean_classgroup_df)]:  # 각 막대 접근
    height = p.get_height()
    axes[0].text(
        x=p.get_x() + p.get_width()/2,  # 막대 중앙
        y=height/2,                     # 막대 중앙에 표시
        s=f"{height:.1f}점",              # 값
        ha='center', va='center',       # 가운데 정렬
        color='white', fontsize=12
    )

total_classgroup_df = df.groupby('class')['total'].mean().reset_index()
# 두 번째 subplot (예시: 동일 데이터, 다른 스타일)
sns.barplot(data=total_classgroup_df, x='class', y='total', 
            hue='class', palette='pastel', alpha=0.6, ax=axes[1])
axes[1].set_xticks(range(len(mean_classgroup_df)))
axes[1].set_xticklabels([x_labels[i] for i in mean_classgroup_df['class']])
axes[1].set_xlabel("")
axes[1].set_ylabel("점수")
axes[1].legend_.remove()  # 범례 제거
axes[1].set_title("총 점수 비교")

for p in axes[1].patches[:len(total_classgroup_df)]:  # 각 막대 접근
    height = p.get_height()
    axes[1].text(
        x=p.get_x() + p.get_width()/2,  # 막대 중앙
        y=height/2,                     # 막대 중앙에 표시
        s=f"{height:.1f}점",              # 값
        ha='center', va='center',       # 가운데 정렬
        color='white', fontsize=12
    )

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="white")
plt.rc('font', family='Malgun Gothic')   # 한글 폰트 설정
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 깨짐 방지

corr = df.drop(['name', 'class', 'mean', 'total'], axis=1).corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, cmap=cmap, mask=mask, vmax=1, center=0.5, fmt='.2f',
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)

plt.title('과목별 비교', fontsize=14, fontweight='bold')
plt.xlabel('', fontsize=12)
plt.ylabel('', fontsize=12)

plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter, MaxNLocator


sns.set_theme(style="white")
plt.rc('font', family='Malgun Gothic')   # 한글 폰트 설정
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 깨짐 방지

cols = df.drop(['class', 'name', 'mean', 'total'],axis=1).columns



for ic in cols:
    # jc 리스트에서 ic 제외
    jc_list = [jc for jc in cols if jc != ic]
    n_subplots = len(jc_list)
    
    # 한 Figure 안에 subplot 생성 (1행, n_subplots 열)
    fig, axes = plt.subplots(1, n_subplots, figsize=(5*n_subplots, 5))
    if n_subplots == 1:
        axes = [axes]  # subplot이 1개일 때 list로 감싸주기
    
    
    for ax, jc in zip(axes, jc_list):
        x = df[ic]
        y = df[jc]

        sizes = 200
        sns.scatterplot(x=x, y=y, s=sizes, color=".15", ax=ax, hue=y, palette="viridis")
        sns.kdeplot(x=x, y=y, levels=20, color="b", linewidths=1, ax=ax)

        m, b = np.polyfit(x, y, 1)
        plot_x = range(x.min(), 100)
        plot_y = m*plot_x + b
        ax.plot(plot_x, plot_y, color="red", linewidth=2)  # 같은 ax에 추가
        
        ax.set_title(f'{ic} vs {jc}', fontsize=12, fontweight='bold')
        ax.set_xlabel(ic)
        ax.set_ylabel(jc)
        ax.legend('')
        ax.set_xticks([tick for tick in ax.get_xticks() if tick <= 100 and tick > 0])
        ax.set_yticks([tick for tick in ax.get_yticks() if tick <= 100 and tick > 0])
        ax.xaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{int(x)}점'))
        ax.yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{int(x)}점'))
        ax.grid(True)

        threshold = 0.3
        if abs(m) > threshold:
            for spine in ax.spines.values():  # 모든 테두리(spine) 접근
                spine.set_edgecolor('red')
                spine.set_linewidth(2)  # 두께 조절
            ax.set_facecolor('#dddddd')
            ax.set_title(label=f'{ic} vs {jc}', fontsize=20, fontweight='bold')
        
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter, MaxNLocator
from scipy import stats

sns.set_theme(style="white")
plt.rc('font', family='Malgun Gothic')   # 한글 폰트 설정
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 깨짐 방지

cols = df.drop(['class', 'name', 'mean', 'total'],axis=1).columns



for i, ic in enumerate(cols):
    # jc 리스트에서 ic 제외
    jc_list = [jc for j, jc in enumerate(cols) if j > i]
    n_subplots = len(jc_list)

    if(n_subplots == 0):
        continue
    
    # 한 Figure 안에 subplot 생성 (1행, n_subplots 열)
    fig, axes = plt.subplots(1, n_subplots, figsize=(5*n_subplots, 5))
    if n_subplots == 1:
        axes = [axes]  # subplot이 1개일 때 list로 감싸주기
    
    
    for ax, jc in zip(axes, jc_list):
        x = df[ic]
        y = df[jc]

        sizes = 200
        sns.scatterplot(x=x, y=y, s=sizes, color=".15", ax=ax, hue=y, palette="viridis")
        sns.kdeplot(x=x, y=y, levels=20, color="b", linewidths=1, ax=ax)

        m, b = np.polyfit(x, y, 1)
        plot_x = range(x.min(), 100)
        plot_y = m*plot_x + b
        ax.plot(plot_x, plot_y, color="red", linewidth=2)  # 같은 ax에 추가
        
        ax.set_title(f'{ic} vs {jc}', fontsize=14, fontweight='bold')
        ax.set_xlabel(ic)
        ax.set_ylabel(jc)
        ax.legend('')
        ax.set_xticks([tick for tick in ax.get_xticks() if tick <= 100 and tick > 0])
        ax.set_yticks([tick for tick in ax.get_yticks() if tick <= 100 and tick > 0])
        ax.xaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{int(x)}점'))
        ax.yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{int(x)}점'))
        ax.grid(True)

        threshold = 0.3
        if abs(m) > threshold:
            for spine in ax.spines.values():  # 모든 테두리(spine) 접근
                spine.set_edgecolor('blue')
                spine.set_linewidth(2)  # 두께 조절
            ax.set_facecolor('#eeeeee')
            ax.set_title(label=f'{ic} vs {jc}', fontsize=16, fontweight='bold')

        r, p_value = stats.pearsonr(x, y)

        if(p_value <= 0.05):
            for spine in ax.spines.values():  # 모든 테두리(spine) 접근
                spine.set_edgecolor('red')
                spine.set_linewidth(5)  # 두께 조절
            ax.set_facecolor('#dddddd')
            ax.set_title(label=f'{ic} vs {jc}', fontsize=20, fontweight='bold', color="red")
        
        if(plot_y[0] > plot_y[-1]):
            ax.text(plot_x[0], plot_y[0], f"상관계수:{r:.2f}\n p-value:{p_value:.2f}", fontsize=12, color='red')
        else:
            ax.text(plot_x[-1], plot_y[-1], f"상관계수:{r:.2f}\n p-value:{p_value:.2f}", fontsize=12, color='red')
            

        
    plt.tight_layout()
    plt.show()
```
